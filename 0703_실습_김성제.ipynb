# 1. Exercise 105: Reverse Order

```python
li = []
while(True):
    try:
        a = int(input('값을 입력하세요, 0을 누르면 종료됩니다. '))
        if a is 0:
            break
        li.append(a)
    except:
        print("숫자를 입력해주세요.") 

print("종료, list: ", li)
for i in li[::-1]:
    print(i)
```

# 2. Exercise 107: Avoiding Duplicates 

```python
li2 = []
while(True):
    a = input("값을 입력하세요, 빈 값을 입력하시면 종료됩니다: ")
    if a is "":
        break
    li2.append(a) if a not in li2 else next
for i in li2:
    print(i)
```

# 3. Exercise 108: Negatives, Zeros and Positives

```python
li_p = []
li_z = []
li_n = []
input_list = []
while(True):
    try:
        a = input('값을 입력하세요, 빈 값을 입력하시면 종료됩니다: ')
        if a == "":
            break
        b = int(a)
        input_list.append(b)
        if b > 0:
            li_p.append(b)
        elif b == 0:
            li_z.append(b)
        elif b < 0:
            li_n.append(b)
    except:
        print("숫자를 입력하세요.")
print(input_list)

f_list = []
for i in li_n:
    f_list.append(i)
for i in li_z:
    f_list.append(i)
for i in li_p:
    f_list.append(i)

print(f_list)
```

# 4. Exercise 110: Perfect Numbers

```python
perfect_num = []
for i in range(1, 10001):
    i_mod = []
    for j in range(1, i):
        if i % j == 0:
            i_mod.append(j)
    if sum(i_mod) == i:
        print(i, " is ", True)
        perfect_num.append(i)
    else:
        print(False)
print(perfect_num)
```

# 5. Exercise 113: Formatting a List

```python
def los(l_o_s): # list_of_strings
    if len(l_o_s) == 1:
        print(l_o_s[0])
    elif len(l_o_s) == 2:
        print(l_o_s[0], " and ", l_o_s[1])
    else:
        strings = l_o_s[0]
        for i in l_o_s[1:len(l_o_s)]:
            if i == l_o_s[len(l_o_s)-1]:
                strings = strings + ' and ' + i
            else:
                strings = strings + ', ' + i
        print(strings)
```

# 7. Exercise 134: Unique Characters 

```python
a = input("값을 입력하세요: ")
nou = []
for i in a:
    if i not in nou:
        nou.append(i)
    else:
        next

print(nou)
print(len(nou))
```

# 8. Exercise 135: Anagrams

```python
f_w = input("첫번째 값을 입력하세요: ")
s_w = input("두번째 값을 입력하세요: ")

fw = []
sw = []

fw.extend(f_w)
sw.extend(s_w)

fw.sort()
sw.sort()

if fw == sw:
    print(f_w, " 와 ", s_w ," 는 anagram입니다.")
else:
    print(f_w, " 와 ", s_w ," 는 anagram이 아닙니다.")
```
